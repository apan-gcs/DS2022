A Jupyter notebook to calculate final numerical and letter grades for a formatted .xlsx gradebook.
Reads in assignment grades and types automatically.
Outputs "final_gradebook.xlsx".

```python
import pandas as pd
```

```python
# Import each of the sheets
# df is the main gradebook
# assignments includes the max score for each assignment
df = pd.read_excel('8 - fakegradebook.xlsx', sheet_name = 0)
assignments = pd.read_excel('8 - fakegradebook.xlsx', sheet_name = 1)
types = pd.read_excel('8 - fakegradebook.xlsx', sheet_name = 2)
letter = pd.read_excel('8 - fakegradebook.xlsx', sheet_name = 3)
df
```

```python
types['Type']
```

```python
# Weight each assignment grade out of its maximum -> [0, 1]

for i in assignments.index:
    col = df.filter(like = str(assignments["Label"][i]), axis=1)
    df[col.columns[0]] = df[col.columns[0]] / assignments["Max"][i]

df
```

```python
# Create a cumulative assignment grades
for category in types['Type']:
    filtered = df.filter(like = category, axis = 1)
    df[category] = filtered.mean(axis=1)
df
```

```python
# Weight each assignment type
df['Cumulative'] = 0.0
for i in types.index:
    df['Cumulative'] += df[types["Type"][i]] * types["Weight"][i]
```

```python
# Assign
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.merge_asof.html
#final = pd.merge_asof(df, letter, left_on='Cumulative', right_on='Threshold')

# Converting Threshold to float and accounting for rounding
letter['Threshold'] = letter['Threshold'] - 0.5
df = df.fillna(0)
final = pd.merge_asof(df.sort_values('Cumulative').round(2), letter.sort_values('Threshold'), left_on='Cumulative', right_on='Threshold')
final
```

```python
# Sort sheet by last name and drop unnecessary columns
final.sort_values('Last Name', inplace=True)
final.drop(types['Type'], axis=1, inplace=True)
final.drop(['Threshold'], axis=1, inplace=True)
final
```

```python
# Plot using built-in plotting methods

# First, drop the the 0 / anything lower than a 50
plot = final.drop(final[final['Cumulative'] < 50.0].index, axis=0)

plot['Cumulative'].plot(kind='hist', title="Grade Distribution", xlim=(82,99), rwidth=0.9, yticks=[0,1,2,3],
                        bins=range(82,99,1)).set_xlabel("Cumulative Score")
```

```python
# Save as an Excel sheet (without the index)
final.to_excel('final_gradebook.xlsx', index=None)
```
